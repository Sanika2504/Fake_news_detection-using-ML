```python
def clean_row(row):
    row = row.lower()
    row = re.sub('[^a-zA-Z]',' ',row)
    token = row.split()
    news = [wnl.lemmatize(word) for word in token if not word is stopwords]
    cleannes_news = ' '.join(news)
```
